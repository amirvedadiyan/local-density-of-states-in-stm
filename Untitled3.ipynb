## $Plot \;\frac{dI/dV}{I/V}\; by\; V$

#### Import Needed Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Read From txt file

```python
V_I = np.loadtxt('spec_data.txt',dtype='float')
V = [v for v,i in V_I]
I = [i for v,i in V_I]
```

#### Calculate $dI,dV,\bar{I},\bar{V}$

```python
n = len(V_I) - 1
result = np.zeros(shape=(n,2))
for m_index in range(n):
    dI = I[m_index+1]-I[m_index]
    dV = V[m_index+1]-V[m_index]
    Ibar = (I[m_index+1]+I[m_index])/2
    Vbar = (V[m_index+1]+V[m_index])/2
    x = Vbar
    y = (dI/dV)/(Ibar/Vbar)
    result[m_index] = [x,y]
```

#### Plot Figure

```python
plt.plot([x for x,y in result],[y for x,y in result])
plt.ylabel('(dI/dV)/(I/V)')
plt.xlabel('V')
plt.show()
```
